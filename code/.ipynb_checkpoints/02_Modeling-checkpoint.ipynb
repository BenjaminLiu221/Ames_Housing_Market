# Modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
import pickle
np.random.seed(43)

%matplotlib inline
```

## Import Scaled Training and Validation Data

```python
X_train = pd.read_csv('../data/X_train_sc.csv', header=None)
X_test = pd.read_csv('../data/X_test_sc.csv', header=None)
y_train = pd.read_csv('../data/y_train.csv', header=None)
y_test = pd.read_csv('../data/y_test.csv', header=None)
```

## Observe the Naive Baseline to beat with our model

```python
np.mean(y_train)
```

## Instantiate our model

```python
lr = LinearRegression()
```

```python
lasso = LassoCV()
```

```python
ridge = RidgeCV()
```

## Fit our model

```python
lr.fit(X_train, y_train)
```

```python
lasso.fit(X_train, y_train.values.ravel())
```

```python
ridge.fit(X_train, y_train)
```

## Compare the model on train and test splits
- Compare on $R^2$ scores to assess model underfit or overfit.
- $R^2$ scores represent the percentage of variance that is explained by the model.

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

- Linear regression model performed well on train split and underperformed on test split indicating extreme overfitting of our model. It didn't predict well on unseen data. There are too many features and will need to do some feature selection or feature engineering.

```python
lasso.score(X_train, y_train)
```

```python
lasso.score(X_test, y_test)
```

- Lasso model performed fairly close between train and test splits with a high percentage of variance explained.

```python
ridge.score(X_train, y_train)
```

```python
ridge.score(X_test, y_test)
```

- Ridge model performed fairly close between train and test splits with a high percentage of variance explained.

```python
pred = lasso.predict(X_test)
```

```python
r2_score(y_test, pred)
```

## RMSE
- Calculate RMSE for model and naive baseline
- RMSE represents on average the square root of the error between actual values and predicted values
- Compare the RMSE of model and the RMSE of naive baseline and every model's RMSE thereafter to the RMSE of the previous model

```python
mean_squared_error(y_test, pred)**0.5
```

```python
naive_baseline = np.array([np.mean(y_train)] * len(pred))
```

```python
mean_squared_error(y_test, naive_baseline)**0.5
```

## Naive Baseline vs Model Predictions
- Plot Naive Baseline and Model Predictions separately vs True Values

```python
plt.figure(figsize=(20,10))
plt.scatter(pred, naive_baseline)
plt.ylabel('Naive Baseline', fontsize=25)
plt.xlabel('True Values', fontsize=25)
plt.plot([0, 500000], [0, 500000], color='r')
plt.title('Naive Baseline vs True', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(['Truth of Predictions'], fontsize=20);
```

The line is where the predicted was actually the true value. The y distance from a point to the line represents the error. In other words, if the point is above the line, the predicted was greater than the true value and if the point is below the line, the predicted value was less than the true value. 

```python
plt.figure(figsize=(20,10))
plt.scatter(pred, y_test)
plt.ylabel('Predicted Values', fontsize=25)
plt.xlabel('True Values', fontsize=25)
plt.plot([0, 500000], [0, 500000], color='r')
plt.title('Predicted vs True', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(['Truth of Predictions'], fontsize=20);
```

The line is where the predicted was actually the true value. The y distance from a point to the line represents the error. In other words, if the point is above the line, the predicted was greater than the true value and if the point is below the line, the predicted value was less than the true value. 

## Import Columns list

```python
with open('../assets/columns.pkl', 'rb') as f:
    columns = pickle.load(f)
```

## Align columns with the $\beta$ coefficients

```python
betas = pd.DataFrame(lasso.coef_, index=columns)
```

## Rename the column and create an absolute values column for sorting

```python
betas.columns = ['weights']
```

```python
betas['abs_w'] = betas.weights.abs()
```

## Observe the $\beta_0$
- $\beta_0$ or intercept, is the base prediction when the value of our predictor is 0.

```python
beta_0 = lasso.intercept_
beta_0
```

## Create a bar plot of the top 16 coefficients by absolute value of their weights

```python
weights = betas.sort_values('abs_w', ascending=False)['weights'].head(16)
```

```python
labels = weights.index
```

```python
plt.figure(figsize=(20,10))
plt.barh(labels, weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.title(f'Most Important Features and Weights\ny intercept: ${beta_0:0.02f}', fontsize=42)
plt.tight_layout()
```

## Interpret the Coefficients

The feature with $\beta$ coefficient of the greatest weight was *overall_qual* because the quality of a home should be one of the most important predictor in sale price, if not the highest. The features pertaining to different floors of the home and their areas of square footage had great weights. This makes sense because square footage is an important predictor in the sale price of a home. This is supported by the fact that number of rooms and number of baths had great weight. The greater the number of rooms and baths, the greater the square footage. The mass veneer area is one of the top coefficients in terms of weight because walls built have square footage which is an important predictor for sale price. The years that homes were remodeled had reasonable weight depending on the decade of the year.
